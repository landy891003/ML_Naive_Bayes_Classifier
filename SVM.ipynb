```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import csv
from numpy import genfromtxt

dataPath='TESTDAT.csv'
data=genfromtxt(dataPath,delimiter=',')
tarPath='TESTTAR.csv'
target=genfromtxt(tarPath,delimiter=',')

X_train=data
y_train=target
```

```python
from sklearn.model_selection import train_test_split
```

```python
dataPath='TRAINDAT.csv'
data=genfromtxt(dataPath,delimiter=',')
tarPath='TRAINTAR.csv'
target=genfromtxt(tarPath,delimiter=',')

X_test=data
y_test=target
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='poly', probability=True)
```

```python
svm.fit(X_train_std,y_train)
```

```python
svm.predict(X_test_std)
```

```python
y_test
```

```python
svm.predict_proba(X_test_std)
```

```python
print('Training Set Score:',svm.score(X_train_std,y_train))
print('Test Set Score:',svm.score(X_test_std,y_test))
```
