```python
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
import csv
from numpy import genfromtxt
```

```python
dataPath='check_all.csv'
data=genfromtxt(dataPath,delimiter=',')
tarPath='check2_all.csv'
target=genfromtxt(tarPath,delimiter=',')

X_train,X_test,y_train,y_test=train_test_split(data,target,test_size=0.3)

#testDataPath='HF_LVEF.csv'
#test_data=genfromtxt(testDataPath,delimiter=',')
#testTarPath='HF2_LVEF.csv'
#test_target=genfromtxt(testTarPath,delimiter=',')

#X_train=data
#y_train=target

#X_test=test_data
#y_test=test_target
```

```python
modelg=GaussianNB()
modelg.fit(X_train,y_train)
modelg.predict(X_test)
```

```python
modelg.predict_proba(X_test)
```

```python
print('Training Set Score:',modelg.score(X_train,y_train))
print('Test Set Score:',modelg.score(X_test,y_test))
```
